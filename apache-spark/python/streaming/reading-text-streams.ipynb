### Counting Words from Local Text Stream

```python
import findspark
findspark.init()

from pyspark import SparkContext
from pyspark.streaming import StreamingContext
```

```python
sc = SparkContext('local[2]', 'StreamingWordCount')
```

```python
# create a streaming context from the SparkContext
# interval defines how often batches will be generated
interval = 1
ssc = StreamingContext(sc, interval)
```

```python
lines = ssc.socketTextStream('localhost', 8080)
```

```python
words = lines.flatMap(lambda line: line.split(' '))
pairs = words.map(lambda word: (word, 1))
word_counts = pairs.reduceByKey(lambda num1, num2: num1 + num2)
```

```python
word_counts.pprint()
```

```python
ssc.start()
```
